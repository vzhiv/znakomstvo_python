````python
import random
import pandas as pd
import seaborn as sns
lst = ['robot'] * random.randint(5,15)
lst += ['human'] * random.randint(5,15)
lst += ['alien'] * random.randint(5,15)
lst += ['angel'] * random.randint(5,15)
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
````

````python
data.loc[data['whoAmI'] == 'robot', 'robot'] = True
data.loc[data['whoAmI'] != 'robot', 'robot'] = False
data.loc[data['whoAmI'] == 'human', 'human'] = True
data.loc[data['whoAmI'] != 'human', 'human'] = False
data.loc[data['whoAmI'] == 'alien', 'alien'] = True
data.loc[data['whoAmI'] != 'alien', 'alien'] = False
data.loc[data['whoAmI'] == 'angel', 'angel'] = True
data.loc[data['whoAmI'] != 'angel', 'angel'] = False
data[['robot', 'human', 'alien', 'angel']]
````

````python
sns.histplot(data = data, x='robot')
````
