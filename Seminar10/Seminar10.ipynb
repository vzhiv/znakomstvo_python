1. Создать новый столбец в таблице с
пингвинами, который будет отвечать за
показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)


```python
penguins.loc[penguins['bill_length_mm'] > 42, 'bill_lenght_group'] = 'long'
penguins.loc[(penguins['bill_length_mm'] <= 42) & (penguins['bill_length_mm'] >= 35), 'bill_lenght_group'] = 'middle'
penguins.loc[penguins['bill_length_mm'] < 35, 'bill_lenght_group'] = 'short'
penguins.head()
```

1. Изобразить гистограмму по flipper_length_mm
